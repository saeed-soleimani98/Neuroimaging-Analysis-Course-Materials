> ### *Reading Binary Numbers*

- *Reading binary numbers follows a similar principle to reading decimal numbers, except each position represents a power of 2 instead of a power of 10.*

- *Therefore, the rightmost digit corresponds to 2⁰ (which equals 1), the next digit to 2¹ (which equals 2), then 2² (4), and so on.*

- *For example, consider the binary number `1011` . From right to left:*

     > *The first (rightmost) digit is 1 × 2⁰ = `1`.*

     > *The second digit is 1 × 2¹ = `2`.*

     > *The third digit is 0 × 2² = `0`.*

     > *The fourth (leftmost) digit is 1 × 2³ = `8`.*

     > *Add these values to find the total in decimal. Hence, `1 + 2 + 0 + 8 = 11` .*

- *This process is the foundation of how to read binary numbers and see what they represent in our usual decimal format.*

> ### *Converting Decimal to Binary*

- *Converting a decimal number to binary can be done using repeated division by 2. The remainders form the bits of the binary number.*

- *Steps to convert decimal 11 to binary:*

     > *Divide 11 by 2 → quotient 5, remainder 1.*
    
     > *Divide 5 by 2 → quotient 2, remainder 1.*
 
     > *Divide 2 by 2 → quotient 1, remainder 0.*

     > *Divide 1 by 2 → quotient 0, remainder 1.*

- *Read the remainders from the last to the first. Hence, the binary number is 1011.*

# **Fundamental Data Types in Python**

*Everything in Python is an **object**.


## *What is an Object?*

*In Python, everything is an object. Numbers, strings, lists, functions, classes, and even modules are objects.*

*An object has three core properties:*

*1.  **Identity** – where it lives in memory (a unique reference).*

*2.  **Type** – what kind of object it is (its class).*

*3.  **Value** – the data it contains.*

````python

x = 42
print(id(x))     # identity (memory reference)
print(type(x))   # <class 'int'>
print(x)         # value: 42
````

| Property | Meaning                   | How to inspect |
| -------- | ------------------------- | -------------- |
| Identity | Unique memory reference   | `id(obj)`      |
| Type     | The object's class        | `type(obj)`    |
| Value    | The object's data/content | `print(obj)`   |


## *we can classify data into two main groups:*

> **Primitive (Basic) Types** and **Container (Collection) Types**.


### *Primitive (Basic) Data Types*

| Data Type | Example | Description |
|:-----------|:---------|:-------------|
| **int** | `5`, `-42`, `100000` | Integer (whole numbers, no decimal point) |
| **float** | `3.14`, `-0.001`, `2.0` | Decimal numbers (floating-point values) |
| **complex** | `3 + 4j` | Complex numbers (with real and imaginary parts) |
| **bool** | `True`, `False` | Boolean logic values used in conditions |
| **str** | `"hello"`, `'Python'`, `"123"` | Sequence of Unicode characters (text) |
| **NoneType** | `None` | Represents “nothingness” or the absence of a value |

##### *Notes:*
- *`int` and `float` are **numeric types** — used for arithmetic operations.*

- *`bool` is actually a subclass of `int` (`True == 1`, `False == 0`).*

- *`str` objects are **immutable** — you cannot change them after creation.*

- *`None` is a **special singleton object** that represents “no value”.*


##### *Container Data Types*

*These are **objects that hold multiple values** — sometimes of different types.*

*They form the foundation of data structures in Python.*

| Data Type | Example | Description |
|:-----------|:----------|:-------------|
| **list** | `[1, 2, 3, "hello"]` | Ordered, **mutable**, allows duplicates, dynamic arrays |
| **tuple** | `(1, 2, 3)` | Ordered, **immutable**, allows duplicates |
| **set** | `{1, 2, 3}` | Unordered, **no duplicates**, mutable but elements must be immutable |
| **dict** | `{"a": 1, "b": 2}` | Unordered, **key–value pairs**, keys must be unique |


##### *notes:*
| Concept | Meaning | Example |
|:---------|:---------|:---------|
| **Mutability** | Whether the value of an object can change in memory | Lists ✔︎ mutable; Tuples ✖︎ immutable |
| **Order** | Whether elements maintain insertion order | Lists & Tuples ✔︎ ordered; Sets ✖︎ unordered |
| **Indexing** | Accessing elements by position | Lists, Tuples, Strings → support indexing |

## **for loop**

- ### *What is a for loop?*

    - *A for loop lets you repeat some code for each item in a group (like a list, string, or range of numbers).*

        - *Think of it as:*

            - *"For each thing in this group, do something."*

| Concept             | Example                       |
| ------------------- | ----------------------------- |
| Loop over list      | `for x in [1, 2, 3]:`         |
| Loop over string    | `for char in "hi":`           |
| Loop with range     | `for i in range(5):`          |
| Loop with condition | `if i > 3:` inside the loop   |
| Nested loop         | `for x in ...: for y in ...:` |


````python
numbers = [3, -5, 8, -2, 7, 1, -9, 4, 0, 6]

# mean 
sum_of_numbers = 0
count = 0
for number in numbers:
    sum_of_numbers += number
    count += 1
mean = sum_of_numbers / count
print(f"Mean of numbers: {mean}")
````

````python

numbers = [3, -5, 8, -2, 7, 1, -9, 4, 0, 6]

# max and min
max_number = numbers[0]
min_number = numbers[0]
for number in numbers:
    if number > max_number:
        max_number = number
    if number < min_number:
        min_number = number
print(f"Largest number: {max_number}")
print(f"Smallest number: {min_number}")
````

````python
numbers = [3, -5, 8, -2, 7, 1, -9, 4, 0, 6]

# positive and negetive numbers
positive_count = 0
negative_count = 0
for number in numbers:
    if number > 0:
        positive_count += 1
    elif number < 0:
        negative_count += 1
print(f"Number of positive numbers: {positive_count}")
print(f"Number of negative numbers: {negative_count}")
````

## **while loop**

### *What is a while loop?*

- *A while loop repeats a block of code as long as a condition is True.*

- *It keeps checking the condition before each loop, and runs again and again until the condition becomes False.*

| Use `for` loop when...                    | Use `while` loop when...                               |
| ----------------------------------------- | ------------------------------------------------------ |
| You know how many times to repeat         | You don’t know how many times — depends on a condition |
| You’re iterating over a list, string, etc | You want to repeat while something is true             |


````python
import random

secret_number = random.randint(1, 100)


attempts = 0

while True:
    guess = 101
    attempts += 1
    
    if guess < secret_number:
        print("Too low! Try again.")
    elif guess > secret_number:
        print("Too high! Try again.")
    else:
        print(f"Congratulations! You guessed the number {secret_number} in {attempts} attempts.")
        
    break
````

## **Functions**

- *A function is a block of code that performs a specific task.*

- *We create it once, and we can use (call) it many times.*

- *Think of it like a machine:*

     - *You give it something (input), it does something (processing), and gives you something back (output).*

````python
def fib(n):
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)


  
````
