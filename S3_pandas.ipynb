# **Using `.groupby()`**

## *The groupby() function follows this approach:*

 > *1 - Split: The data is split into groups based on a specified key (column values).*

 > *2 - Apply: A function (e.g., sum, mean, count) is applied to each group separately.*

 > *3 - Combine: The results are combined into a new DataFrame or Series.*

```python
data = {
    "Department": ["HR", "HR", "IT", "IT", "IT", "Finance", "Finance"],
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace"],
    "Salary": [5000, 6000, 7000, 8000, 7500, 9000, 9500],
    "Experience": [2, 4, 6, 8, 5, 10, 12]
}

df = pd.DataFrame(data)
df
```

```python
# Sum of Salaries per Department
df.groupby("Department")["Salary"].sum()
```

```python
# Average Experience per Department
df.groupby("Department")["Experience"].mean()
```

# **Merging and Joining Different Tables**

| Method         | Type                     | Syntax Example                            | Description / Join Type                                |
| -------------- | ------------------------ | ----------------------------------------- | ------------------------------------------------------ |
|  `pd.concat()` | Vertical      | pd.concat(\[df1, df2], axis=0)            | Stack DataFrames vertically (e.g., sessions, subjects) |
|                | Horizontal  | pd.concat(\[df1, df2], axis=1)            | Combine columns side by side (e.g., features)          |
|  `pd.merge()`  | Inner Join               | pd.merge(df1, df2, on='key')              | Keeps only rows with matching keys in both             |
|                | Left Join                | pd.merge(df1, df2, on='key', how='left')  | All rows from df1 + matching rows from df2             |
|                | Right Join               | pd.merge(df1, df2, on='key', how='right') | All rows from df2 + matching rows from df1             |
|                | Outer Join               | pd.merge(df1, df2, on='key', how='outer') | All rows from both, fill missing with NaN              |


## **Using concat() for Joining Tables**

### *Vertical Concatenation (Stacking Rows)*

```python
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "Salary": [5000, 6000, 7000]
})

df2 = pd.DataFrame({
    "ID": [4, 5, 6],
    "Name": ["David", "Eve", "Frank"],
    "Salary": [7500, 8000, 8500]
})

# Concatenating along rows
result = pd.concat([df1, df2], axis=0)
result

```

```python
result1 = pd.concat([df1, df2], ignore_index=True)
result1
```

### *Horizontal Concatenation (Merging Columns)*

```python
df3 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Department": ["HR", "IT", "Finance"]
})

df4 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Experience": [2, 4, 6]
})

```

```python
result = pd.concat([df3, df4], axis=0)
result
```

## **Using `merge()` for Database-Style Joins**

> *`Inner Join` -> Keeps only matching rows*

> *`Left Join` -> Keeps all rows from the left DataFrame*

> *`Right Join` -> Keeps all rows from the right DataFrame*

> *`Outer Join` -> Keeps all rows from both DataFrames*

```python
df5 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "Salary": [5000, 6000, 7000]
})

df6 = pd.DataFrame({
    "ID": [2, 3, 4],
    "Department": ["IT", "Finance", "HR"]
})
```

### *Inner Join (Keeps only matching rows)*

```python
pd.merge(df5, df6, on="ID")
```

### *Left Join (Keep All Rows from Left)*

```python
pd.merge(df5, df6, on="ID", how="left")
```

### *Right Join (Keep All Rows from Right)*

```python
pd.merge(df5, df6, on="ID", how="right")
```

### *Outer Join (Keep All Rows)*

```python
pd.merge(df5, df6, on="ID", how="outer")
```

# **Summarizing DataFrames in Pandas**

| Function                             | Purpose                                          | Example Code                                |
| ------------------------------------ | ------------------------------------------------ | ------------------------------------------- |
| `df.info()`                          | Overview of columns, dtypes, and non-null counts | `df.info()`                                 |
| `df.describe()`                      | Summary stats of numerical columns               | `df.describe()`                             |
| `df.isnull()`                        | Detect missing values (NaN)                      | `df.isnull()`                               |
| `df.isnull().sum()`                  | Count NaNs per column                            | `df.isnull().sum()`                         |
| `df.notnull()`                       | Detect non-missing values                        | `df.notnull()`                              |
| `df.dropna()`                        | Remove rows with NaNs                            | `df.dropna(subset=['motion'])`              |
| `df.fillna(value)`                   | Fill NaNs with a value                           | `df.fillna(0)`                              |
| `df.rename()`                        | Rename columns                                   | `df.rename(columns={'old':'new'})`          |
| `df.duplicated()`                    | Detect duplicated rows                           | `df[df.duplicated()]`                       |
| `df.duplicated().sum()`              | Count duplicated rows                            | `df.duplicated().sum()`                     |
| `df.value_counts()`                  | Count occurrences of each unique value (1D only) | `df['event_type'].value_counts()`           |
| `df.unique()`                        | List of unique values in a column                | `df['subject'].unique()`                    |
| `df.isin([...])`                     | Filter rows based on a list of values            | `df[df['site'].isin(['siteA','siteB'])]`    |
| `df.nunique()`                       | Count of unique values per column                | `df['diagnosis'].nunique()`                 |
| `df.mean()` / `.median()` / `.std()` | Summary stats (mean, median, std deviation)      | `df['motion'].mean()`                       |
| `df.min()` / `df.max()`              | Min/max of column values                         | `df['age'].min()`                           |
| `df.sort_values(by='col')`           | Sort rows by a column                            | `df.sort_values(by='age', ascending=False)` |
| `df.corr()`                          | Correlation between numeric columns              | `df[['age','motion']].corr()`               |
| `df.groupby(...).describe()`         | Summary stats grouped by category                | `df.groupby('diagnosis')['age'].describe()` |
